```python
# I'll load the uploaded A/B test dataset, run the requested analyses (EDA, two-proportion z-test, CIs, chi-square, t-test),
# show key tables and plots, and summarize results. Outputs will be visible below.
import zipfile
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import os

# Statsmodels for proportion tests & power analysis
from statsmodels.stats.proportion import proportions_ztest, proportion_confint, proportion_effectsize
from statsmodels.stats.power import NormalIndPower

# Attempt to locate the uploaded zip file
zip_path = "ab_data.csv.zip"
if not os.path.exists(zip_path):
    raise FileNotFoundError(f"Uploaded file not found at: {zip_path}")

# Read CSV from zip (assumes a single CSV inside)
with zipfile.ZipFile(zip_path) as z:
    # pick the first csv file inside
    csv_files = [f for f in z.namelist() if f.lower().endswith(".csv")]
    if not csv_files:
        raise FileNotFoundError("No CSV file found inside the uploaded ZIP.")
    csv_name = csv_files[0]
    ab_data = pd.read_csv(z.open(csv_name))

# Display basic info
print("Loaded file:", csv_name)
print("\nShape:", ab_data.shape)
print("\nColumns:", list(ab_data.columns))
print("\nFirst 5 rows:")
display(ab_data.head())

# Basic cleaning: lowercase column names
ab_data.columns = [c.strip().lower() for c in ab_data.columns]

# Ensure expected columns exist
expected_cols = {'group','converted'}
missing = expected_cols - set(ab_data.columns)
if missing:
    print(f"\nWarning: Expected columns {missing} not found. Adjusting analysis where possible.")
    
# Convert converted to int if needed
if 'converted' in ab_data.columns:
    ab_data['converted'] = ab_data['converted'].astype(int)

# EDA: group counts and conversion rates
group_summary = ab_data.groupby('group').agg(
    visitors=('group','count'),
    conversions=('converted','sum'),
    conversion_rate=('converted', 'mean')
).reset_index()
group_summary['conversion_rate_pct'] = (group_summary['conversion_rate']*100).round(3)
display(group_summary)

# Two-proportion Z-test (one-tailed: treatment > control)
conv_counts = group_summary.sort_values('group', ascending=True)['conversions'].values
n_counts = group_summary.sort_values('group', ascending=True)['visitors'].values

# Determine which index is control/treatment mapping
# We'll map by name to be safe
try:
    conv_control = int(group_summary[group_summary['group']=='control']['conversions'].values[0])
    n_control = int(group_summary[group_summary['group']=='control']['visitors'].values[0])
    conv_treat = int(group_summary[group_summary['group']=='treatment']['conversions'].values[0])
    n_treat = int(group_summary[group_summary['group']=='treatment']['visitors'].values[0])
except Exception as e:
    # Fallback: assume order is control then treatment
    conv_control, conv_treat = conv_counts
    n_control, n_treat = n_counts

z_stat, p_value = proportions_ztest([conv_treat, conv_control], [n_treat, n_control], alternative='larger')
print(f"\nTwo-proportion Z-test (H1: treatment > control): Z = {z_stat:.4f}, p-value = {p_value:.6f}")

# 95% confidence intervals for each group's conversion proportion (Wilson)
ci_control = proportion_confint(conv_control, n_control, method='wilson')
ci_treat = proportion_confint(conv_treat, n_treat, method='wilson')
print(f"\nControl conversion rate: {conv_control}/{n_control} = {conv_control/n_control:.4f}")
print(f"95% CI (control, Wilson): ({ci_control[0]:.4f}, {ci_control[1]:.4f})")
print(f"\nTreatment conversion rate: {conv_treat}/{n_treat} = {conv_treat/n_treat:.4f}")
print(f"95% CI (treatment, Wilson): ({ci_treat[0]:.4f}, {ci_treat[1]:.4f})")

# Plot conversion rates with errorbars (95% CI)
plt.figure(figsize=(7,4))
x = [0,1]
rates = [conv_control/n_control, conv_treat/n_treat]
yerr_lower = [rates[0]-ci_control[0], rates[1]-ci_treat[0]]
yerr_upper = [ci_control[1]-rates[0], ci_treat[1]-rates[1]]
yerr = np.array([yerr_lower, yerr_upper])
plt.errorbar(x, rates, yerr=yerr, fmt='o', capsize=8)
plt.xticks(x, ['Control','Treatment'])
plt.ylabel('Conversion rate')
plt.title('Conversion Rates with 95% CI (Wilson)')
plt.grid(alpha=0.2)
plt.show()

# Chi-square test: device vs converted (if device exists)
if 'device' in ab_data.columns:
    ct = pd.crosstab(ab_data['device'], ab_data['converted'])
    chi2, chi_p, dof, exp = stats.chi2_contingency(ct)
    print("\nChi-square test (device vs converted):")
    display(ct)
    print(f"chi2 = {chi2:.4f}, p-value = {chi_p:.6f}, dof = {dof}")
else:
    print("\nNo 'device' column found — skipping chi-square test.")

# T-test: session_duration between groups (if exists)
if 'session_duration' in ab_data.columns:
    dur_control = ab_data[ab_data['group']=='control']['session_duration'].dropna()
    dur_treat = ab_data[ab_data['group']=='treatment']['session_duration'].dropna()
    t_stat, t_p = stats.ttest_ind(dur_treat, dur_control, equal_var=False)
    print(f"\nT-test for session_duration (treatment vs control): t = {t_stat:.4f}, p-value = {t_p:.6f}")
    print(f"mean control = {dur_control.mean():.3f}, mean treatment = {dur_treat.mean():.3f}")
else:
    print("\nNo 'session_duration' column found — skipping t-test for session duration.")

# Sample size calculation for detecting observed difference with 80% power (two-sided)
p1 = conv_control / n_control
p2 = conv_treat / n_treat
effect_size = proportion_effectsize(p2, p1)  # using treatment - control
analysis = NormalIndPower()
required_n = analysis.solve_power(effect_size=effect_size, power=0.8, alpha=0.05, alternative='two-sided')
print(f"\nObserved effect size (Cohen h) = {effect_size:.4f}")
print(f"Estimated required sample per group to detect this effect with 80% power (two-sided): {int(np.ceil(required_n))}")

# Save a short results summary CSV to /mnt/data for download if needed
summary_df = group_summary.copy()
summary_df['ci_lower'] = [ci_control[0], ci_treat[0]]
summary_df['ci_upper'] = [ci_control[1], ci_treat[1]]
summary_df.to_csv('ab_test_summary_results.csv', index=False)
print("\nSummary results saved to: ab_test_summary_results.csv")

# Final concise summary printed
print("\n--- Concise Summary ---")
if p_value < 0.05:
    print(f"The test shows treatment has a significantly higher conversion rate than control (Z={z_stat:.3f}, p={p_value:.6f}).")
else:
    print(f"No statistically significant improvement detected for treatment (Z={z_stat:.3f}, p={p_value:.6f}).")
print(f"Control CR = {p1:.3%} (95% CI {ci_control[0]:.3%}–{ci_control[1]:.3%}); Treatment CR = {p2:.3%} (95% CI {ci_treat[0]:.3%}–{ci_treat[1]:.3%}).")
```
